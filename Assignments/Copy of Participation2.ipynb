Write a function that accepts the list 'spam' as an argument and returns a string with all the items separated by a comma and a space, with 'and' inserted before the last item. The function should work with any list.

```python
def functionone(items):
  return ', '.join(items[:-1]) + ', and ' + items[-1]
spam = ['apples', 'bananas', 'tofu', 'cats']
print(functionone(spam))
```
